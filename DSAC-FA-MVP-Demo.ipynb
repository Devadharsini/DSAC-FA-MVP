```python
#This library is used to initialize the dash application.
import dash
# This library is used to add graphs,other visual components.
from dash import dcc
# This library is used to include html tags.
from dash import html
#For data manipulation and mathematical operations.
import pandas as pd
#Reading the csv file.
data = pd.read_csv("TSLA.csv")
#Manipulating the date.
data["Date"] = pd.to_datetime(data["Date"], format="%Y-%m-%d")
```

```python
app = dash.Dash()
app.layout = html.Div(

children=[

html.H1(children="Stock Market Details of Tesla",),
html.P(
children="Analyzing day wise high and low prices of indexes.",
),
dcc.Graph(
figure={
"data": [
{
"x": data["Date"],
"y": data["High"],
"type": "lines",
},
],
"layout": {"title": "Day-wise highest prices of indexes"},
},
),
dcc.Graph(
figure={
"data": [
{
"x": data["Date"],
"y": data["Low"],
"type": "lines",
},
],
"layout": {"title": "Day-wise lowest prices of indexes"},
},
),
dcc.Graph(
figure={
"data": [
{
"x": data["Date"],
"y": data["Close"],
"type": "lines",
},
],
"layout": {"title": "Day-wise closing prices of indexes"},
},
),
dcc.Graph(
figure={
"data": [
{
"x": data["Date"],
"y": data["Open"],
"type": "lines",
},
],
"layout": {"title": "Day-wise opening prices of indexes"},
},
),
] )


app.run_server()
```

```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objects as go
import plotly.express as px

app = dash.Dash()   

def stock_prices():
    fig = go.Figure([go.Scatter(x = data['Date'], y = data['Adj Close'],\
                     line = dict(color = 'firebrick', width = 4), name = 'Tesla')
                     ])
    fig.update_layout(title = ' Adj Closed Prices over time',
                      xaxis_title = 'Dates',
                      yaxis_title = 'Prices'
                      )
    return fig


app.layout = html.Div(id = 'parent', children = [
    html.H1(id = 'H1', children = 'Tesla Stock Market Analysis', style = {'textAlign':'center',\
                                            'marginTop':40,'marginBottom':40}),


        dcc.Graph(id = 'line_plot', figure = stock_prices())
    ]
                     )
def stock_prices1():
    fig = go.Figure([go.Scatter(x = data['Date'], y = data['Adj Close'],\
                     line = dict(color = 'firebrick', width = 4), name = 'Tesla')
                     ])
    fig.update_layout(title = ' Adj Close Prices over time',
                      xaxis_title = 'Dates',
                      yaxis_title = 'Prices'
                      )
    return fig


app.layout = html.Div(id = 'parent', children = [
    html.H1(id = 'H1', children = 'Ford Stock Market Analysis', style = {'textAlign':'center',\
                                            'marginTop':40,'marginBottom':40}),


        dcc.Graph(id = 'line_plot', figure = stock_prices1())
    ]
                     )

app.run_server()
```
